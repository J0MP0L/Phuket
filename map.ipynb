```python
import json
import pandas as pd
import plotly.io as pio
import plotly.express as px
pio.renderers.default = 'browser'
thai=json.load(open("/Users/thebeyonder/Downloads/thailand-provinces.geojson","r"))
```

```python
thai["features"][1]
```

```python
state_id_map = {}
for feature in thai["features"]:
    feature["id"] = feature["properties"]["ID_1"]
    state_id_map[feature["properties"]["NAME_1"].lower().replace(" ","")] = feature["id"]
```

```python
state_id_map.keys()
```

```python
df=pd.read_excel('/Users/thebeyonder/Downloads/3_20250123_Rice visualization.xlsx',sheet_name="Rice")
df.drop(index=21, inplace=True)
df["Province_EN"][45]="bangkokmetropolis"
df["id"] = df["Province_EN"].apply(lambda x: state_id_map[x.lower().replace(" ","")])
```

```python
import numpy as np
df["scaledensity"]=np.log10(df["Density"])
df.head()
```

```python
fig = px.choropleth_mapbox(
    df,
    locations="id",
    geojson=thai,
    color="เนื้อที่เก็บเกี่ยว (ไร่)",
    hover_name="Province_TH",
    hover_data=["เนื้อที่เก็บเกี่ยว (ไร่)"],
    title="thai Population Density",
    mapbox_style="carto-positron",
    center={"lat": 13.5, "lon": 100.5},
    zoom=5,
    opacity=0.5,
    animation_frame="Region_EN"
)

fig.show()
```
